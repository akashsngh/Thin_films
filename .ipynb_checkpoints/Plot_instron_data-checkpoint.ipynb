## Three test on the same sample  on 08/12/2022

```python
fig, axs = plt.subplots(1,1, figsize =(10,6), constrained_layout= True)
df = []
Initial_position = [47.36,47.36, 47.32] #mm
length = [3,3,3] # cm
t = 1 #mm
b= 4 #mm
files = [r"D:\ESPCI\Experiments\ingstron\ECOFLEX\20221208_153406\20221208_153406_3_1.csv",
         r"D:\ESPCI\Experiments\ingstron\ECOFLEX\20221208_153406\20221208_153406_4_1.csv",
         r"D:\ESPCI\Experiments\ingstron\ECOFLEX\20221208_153406\20221208_153406_5_1.csv"]
for i in range(len(files)):
    df.append(pd.read_csv(files[i],skiprows = 14, encoding='latin-1'))
    df1 = df[i]
    strain = (np.array(df1['Shifted displacement (mm)'])-Initial_position[i])/(length[i]*1e1)
    stress = np.array(df1['Force (N)'])/(t*b*1e-6)
    func = lambda x,p: p * x 
    lim = 0.1
    param, err =  curve_fit(func, strain[(strain<lim)], stress[(strain<lim)])
    axs.plot(strain, stress*1e-3, label = 'exp {}'.format(i))
axs.plot(strain,strain*param[0]*1e-3, linestyle = '--', label = 'linear regime E = {} kPa'.format(round(param[0]*1e-3,3)))#.format(round(param[0],3)))
axs.set_ylabel('stress (kPa)', fontsize =18)
axs.set_xlabel('strain', fontsize = 18)
axs.tick_params(axis="x", labelsize=16)
axs.tick_params(axis="y", labelsize=16)
axs.legend(fontsize = 16)
#plt.savefig('20221208_stress_strain_ingstron.png')
```

```python
import plotly.express as px
import pandas as pd
from scipy.optimize import curve_fit
import numpy as np

df = []
Initial_position = [56.2197,56.252] #mm
length = [3,3,3] # cm
t = 1 #mm
b= 4 #mm
files = [r'D:\ESPCI\Experiments\ingstron\20230104_110633.is_tcyclic_Exports\20230104_110633_1_1.csv',
         r'D:\ESPCI\Experiments\ingstron\20230104_110633.is_tcyclic_Exports\20230104_110633_2_1.csv']

for i in range(len(files)):
    df.append(pd.read_csv(files[i],skiprows = 14, encoding='latin-1'))
    df1 = df[i]
    strain = (np.array(df1['Shifted displacment (mm)'])-Initial_position[i])/(length[i]*1e1)
    stress = np.array(df1['Force (N)'])/(t*b*1e-6)
    func = lambda x,p,q: p * x+q 
    lim = 0.1
    param, err =  curve_fit(func, strain[(strain<lim)], stress[(strain<lim)])
    fig = px.scatter(x=strain, y=stress*1e-3, labels={'x':'strain', 'y':'stress (kPa)'})
    fig.add_scatter(x=strain, y=strain*param[0]*1e-3+param[1]*1e-3, mode='lines', 
                    line={'dash':'dash', 'color':'red'},
                    name='linear regime E = {} kPa'.format(round(param[0]*1e-3,3)))
    fig.update_layout(title='Stress-strain plot',
                      xaxis_title='strain',
                      yaxis_title='stress (kPa)')
    fig.show()
```

```python
import matplotlib.pylab as pl
fig, axs = plt.subplots(1,1, figsize =(14,10), constrained_layout= True)

df = []
length = [2] # cm
t = 1 #mm
b= 4 #mm
n = 8
total_cycles = 28#48

colors = np.array([pl.cm.viridis(np.linspace(0,1,n)),pl.cm.Reds(np.linspace(0,1,n))])
labels = ['sample 1 t = 1 mm, 1 %$s^{-1}$']
files = [r"D:\ESPCI\Experiments\ingstron\ECOFLEX\20240523_111311.is_tcyclic_Exports\20240523_111311_1_1.csv"]
area_under_curve = np.zeros((len(files), total_cycles))
for i in range(len(files)):
    df.append(pd.read_csv(files[i],skiprows = 14, encoding='latin-1'))
    df1 = df[i]
    strain = np.array(df1['Extenso strain (%)'])*1e-2
    stress = np.array(df1['Force (N)'])/(t*b*1e-6)
    cycles = np.array(df1['Nombre total de cycles'])
#lim = 0.1
#param, err =  curve_fit(func, strain[(strain<lim)], stress[(strain<lim)])
    for j in range (n):
        axs.plot(strain[(cycles>=(j)*3) & (cycles<(j+1)*3)], stress[(cycles>=(j)*3) & (cycles<(j+1)*3)]*1e-3, color=colors[i,j], label = 
                labels[i])#, label = labels[i])
        #print(strain[(cycles>=(j)*3) & (cycles<(j+1)*3)].shape)
        ##ax2.plot(strain[(cycles>=(j)*3) & (cycles<(j+1)*3)], stress[(cycles>=(j)*3) & (cycles<(j+1)*3)], color=colors[j])#, label = labels[i])
#axs.plot(strain,strain*param[0]*1e-3, linestyle = '--', label = 'linear regime E = {} kPa'.format(round(param[0]*1e-3,3)))#.format(round(param[0],3)))
axs.set_ylabel('stress (KPa)', fontsize =20)
axs.set_xlabel('strain', fontsize = 20)
axs.tick_params(axis="x", labelsize=18)
axs.tick_params(axis="y", labelsize=18)

#ax2.set_ylabel('stress (kPa)', fontsize =18)
#ax2.set_xlabel('strain', fontsize = 18)
##ax2.tick_params(axis="x", labelsize=16)
##ax2.tick_params(axis="y", labelsize=16)

#axs.legend(fontsize = 16)
##ax2.set_xscale('log')
##ax2.set_yscale('log')
##ax2.set_ylim(0,150)
##ax2.set_xlim(0,0.25)
#plt.savefig(r"D:\ESPCI\seminar\GFP 2023\poster\images\014_20230222_stress_strain_cyclic_ingstron_STODV1_2.pdf")

for i in range(len(files)):
    for j in  range(total_cycles):
        df.append(pd.read_csv(files[i],skiprows = 4, encoding='latin-1'))
        df1 = df[i]
        strain = np.array(df1['Strain AS (%)'])*1e-2
    
        stress = np.array(df1['Stress AS (kPa)'])
        cycles = np.array(df1['Nombre total de cycles'])
        area = trapz(stress[(cycles == (j+1)/2-0.5)],x = strain[(cycles == (j+1)/2-0.5)])
        area_under_curve[i,j] = area
        
    
plt.savefig('20221208_stress_strain_ingstron_RTV_figure_2.png')
```
